```python
import pandas as pd
import numpy as np
```

```python
df_partidas = pd.read_csv('partidas.csv').drop(columns=['Unnamed: 0'])
df_partidas['DateTime UTC'] = pd.to_datetime(df_partidas['DateTime UTC'], errors='coerce')
for i in [1,2]: df_partidas[f'Team{i}RiftHeralds'] = df_partidas[f'Team{i}RiftHeralds'].fillna(0.0)

df = pd.DataFrame()
for i,partida in df_partidas.iterrows():
    # Verificar se existem partidas anteriores dos 2 times na liga
    df_retro_1 = df_partidas[(df_partidas['DateTime UTC']<partida['DateTime UTC'])&
                             ((df_partidas['Team1']==partida['Team1'])|
                              (df_partidas['Team2']==partida['Team1']))]
    df_retro_2 = df_partidas[(df_partidas['DateTime UTC']<partida['DateTime UTC'])&
                             ((df_partidas['Team1']==partida['Team2'])|
                              (df_partidas['Team2']==partida['Team2']))]
    if len(df_retro_1) == 0 or len(df_retro_1) == 0: continue
    dados_retro = {'winrate1':[],'winrate2':[],'Dragons_avg1':[],'Dragons_avg2':[],'Barons_avg1':[],
                   'Barons_avg2':[],'Towers_avg1':[],'Towers_avg2':[],'Kills_avg1':[],'Kills_avg2':[],
                   'Inhib_avg1':[],'Inhib_avg2':[],'Heralds_avg1':[],'Heralds_avg2':[],'Gold_min1':[],
                   'Gold_min2':[]}
    # Para cada partida anterior encontrada do time 1, verificar se esse time era o 1 ou o 2 
    # e aquisitar as estatisticas do respectivo time
    for j,partida_1 in df_retro_1.iterrows():
        if partida_1['Team1'] == partida['Team1']: team = '1'
        else: team = '2'
        if partida_1['WinTeam'] == partida['Team1']: dados_retro['winrate1'].append(1)
        else: dados_retro['winrate1'].append(0)
        dados_retro['Dragons_avg1'].append(partida_1[f'Team{team}Dragons'])
        dados_retro['Barons_avg1'].append(partida_1[f'Team{team}Barons'])
        dados_retro['Towers_avg1'].append(partida_1[f'Team{team}Towers'])
        dados_retro['Kills_avg1'].append(partida_1[f'Team{team}Kills'])
        dados_retro['Inhib_avg1'].append(partida_1[f'Team{team}Inhibitors'])
        dados_retro['Heralds_avg1'].append(partida_1[f'Team{team}RiftHeralds'])
        dados_retro['Gold_min1'].append(partida_1[f'Team{team}Gold']/(partida_1['Gamelength Number']))
    for j,partida_2 in df_retro_2.iterrows():
        if partida_2['Team1'] == partida['Team2']: team = '1'
        else: team = '2'
        if partida_2['WinTeam'] == partida['Team2']: dados_retro['winrate2'].append(1)
        else: dados_retro['winrate2'].append(0)
        dados_retro['Dragons_avg2'].append(partida_1[f'Team{team}Dragons'])
        dados_retro['Barons_avg2'].append(partida_1[f'Team{team}Barons'])
        dados_retro['Towers_avg2'].append(partida_1[f'Team{team}Towers'])
        dados_retro['Kills_avg2'].append(partida_1[f'Team{team}Kills'])
        dados_retro['Inhib_avg2'].append(partida_1[f'Team{team}Inhibitors'])
        dados_retro['Heralds_avg2'].append(partida_1[f'Team{team}RiftHeralds'])
        dados_retro['Gold_min2'].append(partida_1[f'Team{team}Gold']/(partida_1['Gamelength Number']))
    for chave, lista in dados_retro.items():
        dados_retro[chave] = np.mean(lista)
    dados_retro = pd.DataFrame([dados_retro])
    df = pd.concat([df,dados_retro], ignore_index=True)
```
