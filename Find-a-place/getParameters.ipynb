```python
import pandas as pd
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
import re

options = webdriver.ChromeOptions()
options.add_argument("--start-maximized")
options.add_experimental_option("detach", True)
driver = webdriver.Chrome(service=Service(r"C:\python312\chromedriver.exe"), options=options) 

df_apts = pd.read_csv('links_apartamentos.csv')   
to_do = list(df_apts.loc[df_apts['valor'].isnull()])

n = 1
for i, row in df_apts.iterrows():
    if pd.isna(row['valor']):
        porcentagem = df_apts['valor'].notna().sum()*100/len(df_apts)
        if porcentagem >= n:
            print(f"{porcentagem:.2f}% do total.")
            n += 1
    else: continue
    link = row['link']

    driver.get(link)
    time.sleep(5)

    # Extrai os parâmetros desejados (preço, quartos, banheiros, etc.)
    try: valor = driver.find_element(By.XPATH, '//*[@id="__next"]/div/div[2]/div/div/div/div[4]/main/section/div/div[2]/section/div/div[1]/div[1]/div[1]/h3').text[3:]
    except: 
        if 'indisponivel' in driver.current_url:
            df_apts.drop(i, inplace=True)
            continue
    condominio = driver.find_element(By.XPATH, '//*[@id="__next"]/div/div[2]/div/div/div/div[4]/main/section/div/div[2]/section/div/div[1]/div[2]/li/div[2]/span').text[3:]
    iptu = driver.find_element(By.XPATH, '//*[@id="__next"]/div/div[2]/div/div/div/div[4]/main/section/div/div[2]/section/div/div[1]/div[2]/div/li/div[2]/span').text
    quartos = driver.find_element(By.XPATH, '//*[@id="description-house-loaded-details"]/div/div[1]/div/div/div[1]/div/div/p').text.split(' ')[0]
    tamanho = driver.find_element(By.XPATH, '//*[@id="description-house-loaded-details"]/div/div[1]/div/div/div[2]/div/div/p[1]').text.split(' ')[0]
    vagas = driver.find_element(By.XPATH, '//*[@id="description-house-loaded-details"]/div/div[1]/div/div/div[4]/div/div/p').text.split(' ')[0]
    banheiros = driver.find_element(By.XPATH, '//*[@id="description-house-loaded-details"]/div/div[1]/div/div/div[5]/div/div/p').text.split(' ')[0]
    andar = driver.find_element(By.XPATH, '//*[@id="description-house-loaded-details"]/div/div[1]/div/div/div[6]/div/div/p').text
    mobilia = driver.find_element(By.XPATH, '//*[@id="description-house-loaded-details"]/div/div[1]/div/div/div[7]/div/div/p').text
    try:
        driver.find_element(By.XPATH, '//*[@id="tenant/HouseMedia/STREET"]').click()
    except: 
        df_apts.drop(i, inplace=True)
        continue
    janelas = driver.window_handles
    driver.close()
    driver.switch_to.window(janelas[-1])
    match_loc = re.search(r'(-\d{2}\.\d+,-\d{2}\.\d+)', driver.current_url)
    if match_loc: localizacao = match_loc.group(1)
    else: localizacao = None




    # Adiciona os parâmetros ao DataFrame apt_links
    df_apts.at[i, 'valor'] = valor 
    df_apts.at[i, 'condominio'] = condominio  
    df_apts.at[i, 'iptu'] = iptu  
    df_apts.at[i, 'quartos'] = quartos  
    df_apts.at[i, 'tamanho'] = tamanho  
    df_apts.at[i, 'vagas'] = vagas  
    df_apts.at[i, 'banheiros'] = banheiros  
    df_apts.at[i, 'andar'] = andar  
    df_apts.at[i, 'mobilia'] = mobilia
    df_apts.at[i, 'localização'] = localizacao 
    df_apts.to_csv('links_apartamentos.csv', index=False)

driver.quit()
```
